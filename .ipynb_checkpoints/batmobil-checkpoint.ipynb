```python
import random
random.seed(1234)
```

```python
## Excepciones que usaremos a posteriori
class NotAllowedPilotException(Exception):
    pass
class XLessThanZeroException(Exception):
    pass

## funciones de checking que usaremos durante nuestro 


def check_value_more_than_0(x,name="value"):
    """
    Chequea que el valor sea mayor que cero
    """
    if x<0:
        raise XLessThanZeroException("{0} must be >= 0".format(name))


def check_instance_type(weapon_object,inst_type):
    """
    Chequea el tipo de instancia
    """
    if not isinstance(weapon_object,inst_type):
        raise ValueError("object must be {0} type".format(inst_type.__name__))

```

Creamos armas, jocker, Piloto y batmobil

```python

              
"""
Arma con una cantidad de municion que produce cierto dao
"""
class weapon:
    def __init__(self,ammunition_n,damage=1,name="weapon"):
        check_value_more_than_0(ammunition_n,name)
        check_instance_type(ammunition_n,int)
        self._ammunition = ammunition_n
        self._name = name
        self._damage = damage
    
    @property
    def ammunition(self):
        """
        Propiedad municion que tiene el arma
        """
        return self._ammunition
    @ammunition.setter
    def ammunition(self, value):
        """
        Set municion: siempre mayor que 0
        """
        check_value_more_than_0(value,self._name)
        self._ammunition = value
    @property
    def damage(self):
        """
        Damage que produce el arma
        """
        return self._damage
    
    def launch(self,amount=1):
        """
        Disparar
        """
        self.ammunition -=amount

    def attack(self,amount=1):
        """
        Atacar, a diferencia de lanzar, produce un dao
        """
        self.launch(amount)
        return self.damage * amount

    def __repr__(self):
        return "{0}: {1} ammunition".format(self._name,self._ammunition)
    def __str__(self):
        return self.__repr__()
    
    @classmethod
    def batrocket(cls,ammunition):
        """
        batrocket es un tipo de arma que produce dao 1
        """
        return cls(ammunition,damage=1,name="batrocket")
    @classmethod
    def pem(cls,ammunition):
        """
        PEM es un tipo de arma que produce dao 10
        """
        return cls(ammunition,damage=10,name="pem")
"""
Piloto con un determinado nombre y que cuando dispara tiene una 
determinada probabilidad de acertar en el blanco hit_prob
"""
class Pilot:
    def __init__(self,name,hit_prob=0.75):
        self.name = name.strip().lower()
        self._hit_prob = hit_prob
    @property
    def hit_prob(self):
        return self._hit_prob
    def hit_target_launch(self):
        """
        Lanzar un disparo y si es True, ha acertado en el blanco
        """
        if random.random()>= self.hit_prob:
            return True
        else:
            return False
    def __repr__(self):
        return "Pilot {0} with {1} hit_prob ".format(self.name,\
                                                self.hit_prob)
    def __str__(self):
        return self.__repr__()

        
class batmobil:
    def __init__(self,pilot,bats,pems, gas=250,velocity=0):
        self.pilot = pilot # piloto
        self._bats = bats # armas tipo batrocket
        self._pems = pems # armas tipo pem
        self._velocity = velocity # velocidad inicial
        self._turbo = False   # turbo
        self._gas = gas # gasolina del batmobil
        self.__limit_normal_vel = 250 # limite de velocidad sin turbo
        self.__turbo_vel = 350 # velocidad con turbo
        self._ratio_gas_comp = 0.05 # consumo: ratio_gas_comp * velocidad
        self.__gas_limit = 500 # limite de deposito de gasolina
        self.__attacks = 0 # ataques realizados
        self.__allowed_pilots = ["batman","robin"] # posibles pilotos
        ## chequeos de datos introducidos
        # ver que piloto es tipo Pilot
        check_instance_type(pilot,Pilot)
        # ver que es batman o robin
        self._check_pilot(self.pilot.name)
        # ver que bats es tipo weapon
        check_instance_type(bats,weapon)
        # ver que pems es tipo de weapon
        check_instance_type(pems,weapon)
        
    def _check_pilot(self,pilot):
        """
        Chequea que el piloto es alguno de los permitidos
        """
        if pilot not in self.__allowed_pilots:
            raise NotAllowedPilotException("Pilots allowed to drive batmobil are: batman or robin")
    @property
    def velocity(self):
        """
        Propiedad velocidad
        """
        return self._velocity
    @velocity.setter
    def velocity(self,velocity):
        """
        La velocidad ver que es mayor cero y que es menor que la de turbo.
        Cuando se modifica la velocidad, se consume gasolina
        """
        check_value_more_than_0(velocity,"velocity")
        if ((velocity > self.__limit_normal_vel) & (not self.turbo)):
            raise ValueError("Activate Turbo to go faster")
        elif ((self.turbo) & (velocity< self.__limit_normal_vel)):
            raise ValueError("Deactivate Turbo before decreasing velocity")
        self._velocity = velocity
        self.gas_comsumption()
    
    @property
    def turbo(self):
        """
        Propiedad de si est activado el turbo
        """
        return self._turbo
    @turbo.setter
    def turbo(self,boolean):
        """
        Chequeo de que el turbo sea bool y si se activa poner la velocidad correspondiente
        """
        check_instance_type(boolean,bool)
        self._turbo = boolean
        if self._turbo:
            self.velocity = self.__turbo_vel
        else:
            self.velocity = self.__limit_normal_vel
    
    @property
    def gas(self):
        """
        Propiedad gasolina del coche
        """
        return self._gas
    @gas.setter
    def gas(self,gas):
        """
        Que el gas sea mayor que 0 y no est por encima del limite
        """
        check_value_more_than_0(gas,"gas")
        if ( gas >= self.__gas_limit):
            raise ValueError("Gas limit reached")
        self._gas = gas
    def gas_comsumption(self):
        """
        Consumo de gasolina: vel*ratio_gas_comp
        """
        self.gas-= self.velocity * self._ratio_gas_comp
    
    def set_turbo(self,boolean):
        """
        Poner/Quitar el turbo
        """
        self.turbo = boolean
        
    def gas_fill(self,gas):
        """
        Aadir gasolina
        """
        self._gas+=gas
    
    def acelerate(self,velocity):
        """
        Aumentar/Reducir velocidad del coche
        """
        self.velocity+=velocity
    
    def status(self):
        """
        Devuelve status del coche, piloto y armamento.
        """
        return {"pilot":self.pilot.name,
                "pilot_hit_prob":self.pilot.hit_prob,
                "turbo":self.turbo,
                "velocity":self.velocity,
                "pems": self._pems.ammunition,
                "bats":self._bats.ammunition,
                "gas":self.gas,
                "attacks":self.__attacks
               }
    
    def __repr__(self):
        return "batmobil:{0} armed with {1} batrockets and {2} pem".format(self.pilot.name,\
                                                self._bats.ammunition,\
                                                self._pems.ammunition)
    def __str__(self):
        return self.__repr__()
    
    def charge_rocket(self,ammunition):
        """
        Cambiar municion de batrockets
        """
        self._bats.ammunition+=ammunition
        
    def charge_pem(self,ammunition):
        """
        Cambiar municion de PEM
        """
        self._pem.ammunition+=ammunition
    
    def launch_rockets(self,amount=1):
        """
        Lanzar cohetes
        """
        self._bats.launch(amount)      
        
    def launch_pems(self,amount=1):
        """
        Lanzar PEM
        """
        self._pems.launch(amount)
    
    def attack_rockets(self,amount=1):
        """
        Ataque con los cohetes en funcion de la precisin del piloto
        """
        self.__attacks+=1
        if self.pilot.hit_target_launch():
            return self._bats.attack(amount)
        else:
            return 0
    def attack_pems(self,amount=1):
        """
        Ataque con los PEMs en funcin de la precisin del piloto.
        """
        self.__attacks+=1
        if self.pilot.hit_target_launch():
            return self._pems.attack(amount)
        else:
            return 0

    
    
"""
Clase Jocker, con numero de vidas determinado
"""
class jokker:
    def __init__(self,lifes=100,name="jokker"):
        self._KO = False
        self._weapon = True
        self._lifes = lifes
        self._name = name
        self._attacks= 0
    @property
    def KO(self):
        """
        Propiedad de si est muerto o no
        """
        return self._KO
    @KO.setter
    def KO(self,boolean):
        """
        KO: Chequear que es boolean
        """
        check_instance_type(boolean,bool)
        self._KO = boolean
    @property
    def weapon(self):
        """
        Propiedad de si ha sido atacado por PEM
        """
        return self._weapon
    @weapon.setter
    def weapon(self,boolean):
        """
        Chequeo de si es boolean.
        """
        check_instance_type(boolean,bool)
        self._weapon = boolean
    @property
    def lifes(self):
        """
        Numero de vida del jokker
        """
        return self._lifes    
    @lifes.setter
    def lifes(self,life):
        """
        lifes: chequeo de que sean mayor 0. Si es menor, se pone KO como True (muere)
        """
        try:
            check_value_more_than_0(life,"lifes")
            self._lifes = life
        except XLessThanZeroException:
            self._lifes = 0
        if self._lifes ==0:
            self.KO = True 
            
    def attacked(self,damage=1,dtype="rocket"):
        """
        Si es atacado, cuantas vidas le quita y el tipo de arma. si es pem, se pone weapon= True
        """
        self.lifes-=damage
        if dtype=="pem":
            self.weapon = True
        self._attacks+=1
    @classmethod
    def jokkervillain(cls,lifes=50):
        """
        Devuelve el tipo jokker Villain, que tiene menos vidas que el jocker normal
        """
        return jokker(lifes=50,name="jokker villain")
    def __repr__(self):
        return "{0} with {1} lifes ".format(self._name,\
                                                self._lifes)
    def __str__(self):
        return self.__repr__()
    def status(self):
        """
        Devuelve status del jokker
        """
        return {
            "name":self._name,
            "lifes":self.lifes,
            "weapon":self.weapon,
            "KO":self.KO,
            "attacked_n": self._attacks
        }

    
    
def running_battle(batrocket_n = 100,pem_n = 50,hit_prob=0.8,prob_diff_jockers = 0.5):
    """
    Ejemplo de una pelea entre batman y dos jockers.
    """
    def status(btm,jk1,jk2):
            print("Status batmobil...")
            print(btm.status())
            print("Status jokker...")
            print(jk1.status())
            print("Status jokker villain...")
            print(jk2.status())
    # inicializamos armas
    # batrocket
    bats = weapon.batrocket(batrocket_n)
    # pem
    pems = weapon.pem(pem_n)
    # inicializamos jocker
    joker1= jokker()
    # inicializamos jocker Villano
    jokerVill = jokker.jokkervillain()
    # inicializamos piloto
    pilot = Pilot("batman",hit_prob=hit_prob)
    # inicializamos batmobit
    btm = batmobil(pilot,bats,pems)
    counts=0
    # mientras estn los dos jokers vivos, mantenemos la jugada
    while not ((joker1.KO) & (jokerVill.KO)):
        # aleatoriamente modificamos la velocidad
        acerelat = random.randint(0,5)
        btm.acelerate(acerelat)
        # aleatoriamente lanzamos entre 0 y 5 rockets
        rocket_n = random.randint(0,5)
        damage_rocket =  btm.attack_rockets(rocket_n)
        pem_n = random.randint(0,1)
        # aleatoriamente lanzamos entre 0 y 1 PEM
        damage_pem =  btm.attack_pems(rocket_n)
        # elegimos aleatoriamente si atacamos al jocker villano o al jocket
        if random.random()> prob_diff_jockers:
            joker1.attacked(damage_rocket,"rocket")
            joker1.attacked(damage_pem,"pem")
        else: 
            jokerVill.attacked(damage_rocket,"rocket")
            jokerVill.attacked(damage_pem,"pem")
        counts+=1
        
        if counts%20==0:
            # cada 20 iteraciones mostramos status
            print("Iterations:{0}".format(counts))
            status(btm,joker1,jokerVill)
    print("Jokers are dead!")
    status(btm,joker1,jokerVill)
```
